```python
from tkinter import *
import tkinter.ttk as ttk
from deep_translator import GoogleTranslator


root = Tk()

root.title("Language Translator Tool")
root.geometry("1100x340")
root.resizable(0, 0)
root.configure(bg="skyblue")


Label(
    root,
    text="Language Translator",
    font="Arial 20 bold",
    bg="skyblue",
    fg="black"
).pack(pady=10)


Label(
    root,
    text="Enter Text:",
    font="arial 12 bold",
    bg="skyblue"
).place(x=50, y=60)

Input_text = Text(
    root,
    font="arial 13",
    height=5,
    wrap=WORD,
    padx=5,
    pady=5,
    width=50
)

Input_text.place(x=50, y=90)


Label(
    root,
    text="Translated Output:",
    font="arial 12 bold",
    bg="skyblue"
).place(x=600, y=60)

Output_text = Text(
    root,
    font="arial 13",
    height=5,
    wrap=WORD,
    padx=5,
    pady=5,
    width=50
)

Output_text.place(x=600, y=90)


languages = {
    'english': 'en',
    'hindi': 'hi',
    'spanish': 'es',
    'french': 'fr',
    'german': 'de',
    'chinese (simplified)': 'zh-CN',
    'chinese (traditional)': 'zh-TW',
    'japanese': 'ja',
    'korean': 'ko',
    'arabic': 'ar',
    'russian': 'ru'
}


Label(
    root,
    text="Source Language",
    font="arial 11 bold",
    bg="skyblue"
).place(x=50, y=210)

source_lang = ttk.Combobox(
    root,
    values=list(languages.keys()),
    width=25,
    font="arial 11"
)

source_lang.place(x=50, y=240)
source_lang.set("english")


Label(
    root,
    text="Target Language",
    font="arial 11 bold",
    bg="skyblue"
).place(x=750, y=210)

dest_lang = ttk.Combobox(
    root,
    values=list(languages.keys()),
    width=25,
    font="arial 11"
)

dest_lang.place(x=750, y=240)
dest_lang.set("hindi")


def Translate():
    try:

        source_selected = source_lang.get().lower().strip()
        target_selected = dest_lang.get().lower().strip()

        if source_selected in languages and target_selected in languages:

            source_code = languages[source_selected]
            target_code = languages[target_selected]

            translator = GoogleTranslator(
                source=source_code,
                target=target_code
            )

            text_to_translate = Input_text.get("1.0", END).strip()

            translated_text = translator.translate(text_to_translate)

            Output_text.delete(1.0, END)
            Output_text.insert(END, translated_text)

        else:
            Output_text.delete(1.0, END)
            Output_text.insert(END, "Please select valid languages")

    except Exception as e:
        Output_text.delete(1.0, END)
        Output_text.insert(END, f"Translation Error:\n{str(e)}")


def Copy():
    translated_data = Output_text.get("1.0", END)

    root.clipboard_clear()
    root.clipboard_append(translated_data)


translate_btn = Button(
    root,
    text="Translate",
    font="arial 12 bold",
    padx=5,
    pady=5,
    bg="orange",
    fg="black",
    activebackground="green",
    command=Translate
)

translate_btn.place(x=500, y=235)

copy_btn = Button(
    root,
    text="Copy",
    font="arial 12 bold",
    padx=5,
    pady=5,
    bg="black",
    fg="white",
    command=Copy
)

copy_btn.place(x=610, y=235)


root.mainloop()
```
